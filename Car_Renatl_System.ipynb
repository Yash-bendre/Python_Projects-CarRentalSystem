# Car Rental Sysytem

```python
cars_showeroom={}
rented_cars={}

#Create function to add cars
def add_cars():
    print("\n---Add Cars To The System---")
    while True:
        car_id=input("Enter The car Id (e.g., 001) :")
        if car_id in cars_showeroom:
            print("Car Id already exist. Try Anthor")
            continue
        brand=input("Enter the car Brand")
        model=input("Enter the cars Model")
        try:
            rate=float(input("Enter the rate per day: "))
        except ValueError:
            print("Invalid Rate")
            continue

        #add cars To the dict
        cars_showeroom[car_id]={
            "Brand":brand,
            "Model":model,
            "Rate_per_day":rate,
            "Isavialable":True
        }

        anthor=input("Add anthor car? (y/n) :").lower()
        if anthor!='y':
            break
        

#create function for disply cars details
def disply_cars():
    print("\n---Available Cars In Showeroom---")
    if not cars_showeroom:
        print("No Cars Avialable, Plese add cars First")
        return
    for car_id,car_details in cars_showeroom.items():
        status="Available For Rent" if car_details["Isavialable"] else "Rented"
        print(f"Car_Id : {car_id} | {car_details["Brand"]} | {car_details["Model"]} | $ {car_details["Rate_per_day"]} | {status}")

#Create function for rent car
def rent_car():
    print("---WELCOME---")
    customer_name=input("Plese Enter Your Name")
    customer_phone=int(input("Plese Enter the phone number"))
    car_id=input("Plese Enter the car Id")

    if car_id in cars_showeroom and cars_showeroom[car_id]["Isavialable"]:
        try:
            day=int(input("Enter the Number Of Days to Rent The Car"))
            if(day<=0):
                print("Day Must Grater Than zero")
        except ValueError:
            print("Invalid Input")

        cost=cars_showeroom[car_id]["Rate_per_day"] * day
        cars_showeroom[car_id]["Isavialable"]=False

        rented_cars[car_id]={
            "Name":customer_name,
            "Phone":customer_phone,
            "Total days":day
        }

        print(f"\n....Renatl Recepit....")
        print(f"Customer_Name: {customer_name}")
        print(f"Customer_Phone: {customer_phone}")
        print(f"Car_Detail: {cars_showeroom[car_id]["Brand"]} | {cars_showeroom[car_id]["Model"]}")
        print(f"Total Days You Rent Car :{day}")
        print(f"Total Amount : ${cost}")
    else:
        print("Car Not Avaliable For The Rent")


#Create function For The Return Car
def return_car():
    car_id=input("Enter the car_id")
    if car_id in rented_cars and not cars_showeroom[car_id]["Isavialable"]:
        try:
            actual_days=int(input("Enter How Many Days You Kept The Car"))
        except ValueError:
            print("Invalid Input")
            return

        rented_info=rented_cars[car_id]
        base_cost=cars_showeroom[car_id]["Rate_per_day"]*actual_days 
        cars_showeroom [car_id]["Isavialable"]=True
        del rented_cars[car_id]


        print("\n....Return Deatails....")
        print(f"Customer: {rented_info["Name"]}")
        print(f"Car_Detail: {cars_showeroom[car_id]["Brand"]} | {cars_showeroom[car_id]["Model"]}")
        print(f"Days Used: {actual_days}")
        print(f"Total Amount Due: ${base_cost}")
    else:
        print("Invalid Id Or Car Wasent Return")


#Create Main Method
def main():
    while(True):
        print("\n-^-^-^-Car Rental System-^-^-^-")
        print("1.----Add Cars To The System----")
        print("2.----Display Cars----")
        print("3.----Rent A Cars----")
        print("4.----Return A Cars----")
        print("5.----Exit----")
        

        choice=int(input("Enter Your Choice"))
        if choice==1:
            add_cars()
        elif choice==2:
            disply_cars()
        elif choice==3:
            rent_car()
        elif choice==4:
            return_car()
        elif choice==5:
            print("Thank You For Using The Car Rental System \nHave A Good Day !!")
            break
        else:
            print("Invalid Choice \nPlese Check Showerrooms Menue !!")
            break


main()
    
     
```
